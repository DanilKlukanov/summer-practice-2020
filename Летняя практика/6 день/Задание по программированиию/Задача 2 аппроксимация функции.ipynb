Импортируем неоходимые библиотеки 

```python
%matplotlib inline
import math
import scipy
from scipy import linalg
import numpy as np
from matplotlib import pylab as plt
```

Задаем функцию f(x) = sin(x / 5) * exp(x / 10) + 5 * exp(-x / 2)

```python
def f(x):
    return math.sin(x / 5.0) * math.exp(x / 10.0) + 5 * math.exp(-x / 2.0)
```

```python
x = np.arange(1, 16, 0.1)
y = list(map(f, x))
```

Отрисуем график функции f(x)

```python
plt.plot(x, y)
```

Функция для создания значений x для систем линейных уравнений

```python
def get_matrix(x_list):
    matrix = []
    n = len(x_list)
    for x in x_list:
        stroke = []
        for i in range(n):
            stroke.append(x ** i)
        matrix.append(stroke)
    return np.array(matrix)
```

Функция для расчета приближенных значений

```python
def aprox_val(w, x_list):
    y = []
    for x in x_list:
        y_val = 0
        for i in range(len(w)):
            y_val += w[i]*(x ** i)
        y.append(y_val)
    return y
```

Находим функцию f1(x) через 2 точки: 1, 15

```python
x_list_2 = [1, 15]
A_2 = get_matrix(x_list_2)
b_2 = np.array(list(map(f, x_list_2)))
w_2 = scipy.linalg.solve(A_2, b_2)
y_2 = aprox_val(w_2, x)
```

Строим график функции f(x) и f1(x)

```python
plt.plot(x, y)
plt.plot(x, y_2)
```

Находим функцию f2(x) через 3 точки: 1, 8, 15

```python
x_list_3 = [1, 8, 15]
A_3 = get_matrix(x_list_3)
b_3 = np.array(list(map(f, x_list_3)))
w_3 = scipy.linalg.solve(A_3, b_3)
y_3 = aprox_val(w_3, x)
```

Строим график функции f(x) и f2(x)

```python
plt.plot(x, y)
plt.plot(x, y_3)
```

Находим функцию f3(x) через 4 точки: 1, 4, 10, 15

```python
x_list_4 = [1, 4, 10, 15]
A_4 = get_matrix(x_list_4)
b_4 = np.array(list(map(f, x_list_4)))
w_4 = scipy.linalg.solve(A_4, b_4)
y_4 = aprox_val(w_4, x)
```

Строим график функции f(x) и f3(x)

```python
plt.plot(x, y)
plt.plot(x, y_4)
```
